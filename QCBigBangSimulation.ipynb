# Quantum Simulation of the Big Bang with Qiskit

Below is a Qiskit code that simulates a quantum-inspired model of the Big Bang, incorporating key cosmological concepts into a quantum circuit.

## Quantum Cosmology Simulation Explained

This code models key aspects of the Big Bang cosmology using quantum computing concepts:

### 1. Initial Singularity
- Starts with a single qubit in superposition: `|0⟩ → (|0⟩ + |1⟩)/√2`
- Represents the quantum fluctuation in the singularity that initiated the Big Bang

### 2. Cosmic Inflation
- Rapid expansion phase implemented by adding entangled qubits
- Exponential growth of quantum complexity
- Simulated through Hadamard gates and CNOT entanglements

### 3. Matter-Antimatter Creation
- Particle creation through rotation gates (RY, RX)
- Represents matter-antimatter pairs emerging from quantum foam
- Asymmetric matter creation through probabilistic operations

### 4. Structure Formation
- Gravity-like interactions via controlled operations (CNOT)
- Formation of cosmic structures through entanglement patterns
- Dark energy effects simulated with phase rotations (RZ)

### 5. Cosmic Evolution Metrics
- **Quantum Entropy**: Measures increasing complexity of the universe
- **Matter Content**: Tracks density of excited states (|1⟩)
- **Anti Matter Content**: Tracks density of |0> states
- **Energy Concentration**: Shows distribution of probability amplitudes

## Simulation Outputs

1. **Animated Evolution**:
   - Real-time visualization of the universe's quantum state
   - Shows probability distribution across all possible states
   - Tracks qubit count, entropy, matter, and anti-matter content

2. **Cosmic Parameter Plots**:
   - Evolution of entropy, matter, anti-matter, and energy over cosmic time
   - Shows phases of inflation, reheating, and structure formation

3. **Final State Distribution**:
   - Histogram of measurement outcomes
   - Represents the quantum state of the "present-day" universe
   - Shows complex entanglement patterns from cosmic evolution

## Scientific Interpretation

While this is a highly simplified model, it demonstrates how quantum systems can capture key cosmological phenomena:

- **Quantum to Classical Transition**: Initial superposition evolves to definite structures
- **Emergent Complexity**: Simple rules generate complex entanglement patterns
- **Irreversibility**: Increasing entropy mirrors the arrow of time
- **Structure Formation**: Gravity-like interactions create correlations

This simulation provides a quantum information perspective on cosmic evolution, illustrating how fundamental quantum processes might underlie the universe's development from singularity to complex structure.

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Big Bang simulation
INITIAL_QUBITS = 1  # Singularity state
FINAL_QUBITS = 8    # Expanded universe state
INFLATION_STEPS = 10
TIME_STEPS = 20
HUBBLE_CONSTANT = 0.15
DARK_ENERGY_STRENGTH = 0.05
MATTER_CREATION_RATE = 0.1

# Store evolution history
evolution_history = []
entropy_data = []
matter_data = []
anti_matter_data = []
energy_data = []
qubit_counts = []

# Create a figure for cosmic evolution
plt.figure(figsize=(15, 10))

# Initialize the quantum universe
current_circuit = QuantumCircuit(INITIAL_QUBITS)
current_circuit.h(0)  # Quantum fluctuation in the singularity
active_qubits = INITIAL_QUBITS  # Track active qubits

# Function to expand the universe (add qubits)
def expand_universe(qc, active_qubits, step):
    if active_qubits < FINAL_QUBITS and step < INFLATION_STEPS:
        # Create a new circuit with additional qubit
        new_qc = QuantumCircuit(active_qubits + 1)

        # Get current statevector as a numpy array
        old_state = Statevector(qc)
        old_state_data = old_state.data

        # Create new state vector with additional qubit
        new_state = np.zeros(2**(active_qubits + 1), dtype=complex)
        new_state[:len(old_state_data)] = old_state_data
        
        # Initialize new circuit with combined state
        new_qc.initialize(new_state, range(active_qubits + 1))
        
        # Entangle new qubit with random existing qubit
        existing_qubit = np.random.randint(0, active_qubits)
        new_qc.h(active_qubits)
        new_qc.cx(existing_qubit, active_qubits)
        
        active_qubits += 1
        return new_qc, active_qubits
    return qc, active_qubits

# Function to apply cosmological evolution
def cosmic_evolution(qc, active_qubits, step):
    # Inflationary period (first few steps)
    if step == 0:
        for qubit in range(active_qubits):
            qc.rx(HUBBLE_CONSTANT * 5, qubit)  # Rapid expansion
    elif step < INFLATION_STEPS:
        print("Inflationary period:Rapid expansion (superposition creation)")
        # Rapid expansion (superposition creation)
        for qubit in range(active_qubits):
            qc.rx(HUBBLE_CONSTANT * 5, qubit)  # Rapid expansion
            qc.ry(HUBBLE_CONSTANT * 5, qubit)  # Rapid expansion
            qc.rz(HUBBLE_CONSTANT * 5, qubit)  # Rapid expansion
    
    # Matter creation era
    elif step < INFLATION_STEPS + 5:
        print("Matter creation era:Particle-antiparticle creation")
        # Particle-antiparticle creation
        for qubit in range(active_qubits):
            if np.random.random() < MATTER_CREATION_RATE:
                qc.ry(np.pi/4, qubit)
                qc.rx(np.pi/4, qubit)
    
    # Structure formation era
    else:
        print("Structure formation era:Gravity-like interactions (entanglement)")
        # Gravity-like interactions (entanglement)
        for i in range(active_qubits - 1):
            for j in range(i+1, active_qubits):
                if np.random.random() < 0.3:
                    qc.cx(i, j)
        
        # Dark energy acceleration
        for qubit in range(active_qubits):
            print("Dark energy acceleration")
            qc.rz(DARK_ENERGY_STRENGTH * step, qubit)
    
    return qc

# Simulate cosmic evolution step-by-step
for step in range(TIME_STEPS):
    print(f"Simulating cosmic time step {step+1}/{TIME_STEPS}...")
    
    # Expand the universe
    current_circuit, active_qubits = expand_universe(current_circuit, active_qubits, step)
    
    # Apply cosmic evolution
    current_circuit = cosmic_evolution(current_circuit, active_qubits, step)
    
    # Simulate the current state
    simulator = Aer.get_backend('statevector_simulator')
    state = Statevector(current_circuit)
    probabilities = state.probabilities()
    
    # Calculate metrics
    entropy = -sum(p * np.log2(p) for p in probabilities if p > 0)
    matter_content = sum(p for i, p in enumerate(probabilities) 
                     if bin(i).count("1") > active_qubits/2)
    anti_matter_content = sum(p for i, p in enumerate(probabilities) 
                     if bin(i).count("0") > active_qubits/2)
    energy_concentration = np.sum(probabilities**2)
    
    # Store data
    entropy_data.append(entropy)
    matter_data.append(matter_content)
    anti_matter_data.append(anti_matter_content)
    energy_data.append(energy_concentration)
    qubit_counts.append(active_qubits)
    evolution_history.append(probabilities.copy())
    
    # Plot current state every 5 steps
    if step % 5 == 0 or step == TIME_STEPS-1:
        plt.figure(figsize=(12, 6))
        states = [f"|{bin(i)[2:].zfill(active_qubits)}⟩" for i in range(2**active_qubits)]
        plt.bar(states, probabilities, color='skyblue')
        plt.title(f'Cosmic Evolution: Step {step} (Qubits: {active_qubits})')
        plt.xlabel('Quantum State')
        plt.ylabel('Probability')
        plt.xticks(rotation=90, fontsize=4)
        plt.ylim(0, 1)
        plt.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()

# Plot cosmic evolution parameters
#plt.figure(figsize=(12, 8))
plt.figure(figsize=(10, 6))
plt.plot(range(TIME_STEPS), entropy_data, 'b-', label='Quantum Entropy (Complexity)')
plt.plot(range(TIME_STEPS), matter_data, 'r-', label='Matter Content')
plt.plot(range(TIME_STEPS), anti_matter_data, 'y-', label='Anti Matter Content')
plt.plot(range(TIME_STEPS), energy_data, 'g--', label='Energy Concentration')
plt.title("Cosmic Evolution Parameters", fontsize=16)
plt.xlabel("Cosmic Time Step", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Plot qubit expansion
plt.figure(figsize=(10, 5))
plt.plot(range(TIME_STEPS), qubit_counts, 'mo-', linewidth=2)
plt.title("Universe Expansion (Qubit Count)", fontsize=16)
plt.xlabel("Cosmic Time Step", fontsize=12)
plt.ylabel("Number of Qubits", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Final state visualization
final_circuit = current_circuit
final_circuit.measure_all()
# Display the quantum circuit
print("\nFinal Quantum Circuit Structure:")
final_circuit.draw(output='mpl', fold=100)

# Simulate final state
simulator = Aer.get_backend('qasm_simulator')
result = execute(final_circuit, simulator, shots=10000).result()
counts = result.get_counts()

plt.figure(figsize=(14, 8))
plot_histogram(counts, title="Final State of the Quantum Universe")
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()

# Print cosmic information
print("\n" + "="*60)
print("QUANTUM COSMOLOGY REPORT: BIG BANG SIMULATION")
print("="*60)
print(f"Initial State: |0⟩ (Quantum Singularity)")
print(f"Final State: {active_qubits}-qubit quantum universe")
print(f"Total Evolution Steps: {TIME_STEPS}")
print(f"Final Quantum Entropy: {entropy_data[-1]:.2f} bits")
print(f"Final Matter Content: {matter_data[-1]:.2%}")
print(f"Final Anti Matter Content: {anti_matter_data[-1]:.2%}")
print(f"Final Energy Concentration: {energy_data[-1]:.2f}")
print(f"Notable Quantum Structures: {len([c for c in counts.values() if c > 500])}")
print(f"Notable Quantum Structures represented as qubit states: {[key for (key, value) in counts.items() if value > 500]}")
print("="*60)
```
